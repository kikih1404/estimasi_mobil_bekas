Jumlah Mobil berdasarkan model

```python

# Mengelompokkan data berdasarkan 'model' dan menghitung jumlah tiap model
models = df.groupby('model')['tax'].count().sort_values(ascending=True).reset_index()

# Mengganti nama kolom 'tax' menjadi 'nomberofcaes'
models = models.rename(columns={'tax': 'nomberofcaes'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['model'], y=models['nomberofcaes'], color='royalblue')
plt.xticks(rotation=60)
```

ukuran mesin

```python
# Mengelompokkan data berdasarkan 'engineSize' dan menghitung jumlah tiap ukuran mesin
engine = df.groupby('engineSize')['tax'].count().sort_values(ascending=True).reset_index()
engine = engine.rename(columns={'tax': 'count'})
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=engine['engineSize'], y=engine['count'], color='royalblue')
```

```python
#distribusi mileage

plt.figure(figsize=(15,5))
sns.distplot(df['mileage'])
```

```python
#distribusi harga mobil

plt.figure(figsize=(15,5))
sns.distplot(df['price'])
```

```python
#seleksi fitur

features = ['year','mileage','tax','mpg','engineSize']
x = df[features]
y = df['price']
x.shape, y.shape
```

```python
#split data training dan data testing

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

```python
#Membuat model regresi liniar

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
```

```python
score = lr.score(x_test, y_test)
print('akurasi model regresi linier = ', score)
```

```python
#Membuat inputan model regresi linier
#terdiri dari year=2019, mileage=5000, tax=145, mpg=50.2, enginesize=2

input_data = np.array([[2019,5000,145,30.2,2]])

prediction = lr.predict(input_data)
print('estimasi harga mobil dalam EUR :', prediction)
```

```python
#membuat save model

import pickle

filename = 'estimasi_mobil.sav'
pickle.dump(lr,open(filename,'wb'))
```
